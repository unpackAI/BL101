### 5ï¸âƒ£  Pool Token
#### Uniswap ğŸ¦„  
>Uniswap is the biggest decentralized crypto exchange. 

Uniswap is an automated liquidity protocol powered by a constant product formula and implemented in a system of non-upgradeable smart contracts on the Ethereum blockchain. It obviates the need for trusted intermediaries, prioritizing decentralization, censorship resistance, and security.

Each Uniswap smart contract, or pair, manages a liquidity pool made up of reserves of two ERC-20 tokens.

Anyone can become a liquidity provider (LP) for a pool by depositing an equivalent value of each underlying token in return for pool tokens. These tokens track pro-rata LP shares of the total reserves, and can be redeemed for the underlying assets at any time.

<center><img src="https://docs.uniswap.org/assets/images/lp-c0b1b03ef921f1325971fa8ab6e9a4f1.jpg" alt="Paramater Settings in Liquidity Pool" width="700"/></center>

Pairs act as automated market makers, standing ready to accept one token for the other. Because the relative price of the two pair assets can only be changed through trading, divergences between the Uniswap price and external prices create arbitrage opportunities. This mechanism ensures that Uniswap prices always trend toward the market-clearing price. Uniswap applies a 0.30% fee to trades.

#### Why do we need Uniswap?
In a nutshell: ğŸ¥œ your ERC-20 token you just created is just a ERC-20 smart contract with some basic information (token name, symbol, decimals, number of tokens). For your token to become *valuable* and *tradeable* or *useable* (in any meaningful way) you need to back it up with "cash", a.k.a. other tokens which are *already* traded and therefore have *already* value. In other words: you need to provide liquiditiy for your token. *That's what we do in this chapter.*

Then in the following chapter we will use the core feature of Uniswap: exchanging our token with other tokens.

### Extract Public Key from Private Key
There is some heavy math behind: *Elliptic Curve Digital Signature Algorithm* ğŸ˜±ğŸ¤¯ which we don't need to understand. ğŸ¤­ 

We only need to know that when you open an account, Ethereum starts with creating a `private key` from which it extracts the `public key` from which it further extracts the `wallet address`: 

> `private key` --> `public key` -> `wallet address`  _(... but it's not working the other way around!)_

```python
# Public Key from private key
# https://stackoverflow.com/a/51954186
from eth_keys import keys
from eth_utils import decode_hex

priv_key_bytes = decode_hex(private_key)
priv_key = keys.PrivateKey(priv_key_bytes)
public_key = priv_key.public_key
print("Your public key: ", public_key)
```

### Extract Wallet Address from Private Key

### Send Money ğŸ’¸
>Of course we could simply use the **Send Button** in your wallet ğŸ¦Š. But where would all the fun be ğŸ¤ª, if we don't code from scratch ğŸ“

**For sending Eth between two blockchain accounts, you need:**   
1. the `private key` of the account you are sending *from* ğŸ‘ˆğŸ» from_account_pk
2. the `wallet address` of the account you are sending *to* ğŸ‘ˆğŸ» to_account
3. how much you want to send

```python
# SEND ETH BETWEEN TWO ACCOUNTS âœ…

#We need: 
#   a) the private key of the account you are sending FROM ğŸ‘ˆğŸ» from_account_pk
#   b) the wallet address of the account you are sending TO ğŸ‘ˆğŸ» to_account

# 1ï¸âƒ£ INPUT DATA
from_account_pk = private_key # ğŸ‘ˆğŸ»  change
to_account = wallet_address2 # ğŸ‘ˆğŸ»  change
amount = 0.1 # ğŸ‘ˆğŸ»  change

# 2ï¸âƒ£ STORE SOME INFOS FOR LATER
from_account = w3.eth.account.privateKeyToAccount(from_account_pk).address
from_account_balance_before = web3.eth.getBalance(from_account)
to_account_balance_before = web3.eth.getBalance(to_account)

# 3ï¸âƒ£ START TRANSACTION
#get the nonce.  Prevents one from sending the transaction twice
nonce = web3.eth.getTransactionCount(from_account)

#build a transaction in a dictionary
tx = {
    'nonce': nonce,
    'to': to_account,
    'value': web3.toWei(amount, 'ether'),
    'gas': 2000000,
    'gasPrice': web3.toWei('50', 'gwei')
}

#sign the transaction
signed_tx = web3.eth.account.sign_transaction(tx, from_account_pk)

#send transaction
tx_hash = web3.eth.sendRawTransaction(signed_tx.rawTransaction)
```

```python
# PRINTOUT TRANSACTION

# convert to Eth
from_account_balance_before_eth = float(web3.fromWei(from_account_balance_before, 'ether')) 
from_account_balance_after_eth = float(web3.fromWei(from_account_balance_after, 'ether')) 
to_account_balance_before_eth = float(web3.fromWei(to_account_balance_before, 'ether')) 
to_account_balance_after_eth = float(web3.fromWei(to_account_balance_after, 'ether')) 

# build table
df = pd.DataFrame(data={"Account Number": [from_account, to_account], "Before (ETH)": [from_account_balance_before_eth, from_account_balance_after_eth], 
                        "After (ETH)": [to_account_balance_before_eth, to_account_balance_after_eth]}, 
                    index=["from:", "to:"], 
                    columns=["Account Number", "Before (ETH)", "After (ETH)"])
df["Difference"] = df["After (ETH)"]-df["Before (ETH)"]

# print stats & table
print("Transaction Hash: ", web3.toHex(tx_hash))
print(f"Amount sent: {amount} Eth")
gas_payed = tx["gas"] * web3.toWei(tx["gasPrice"], 'wei')
print(f"Gas payed: {gas_payed} Wei")
df
```
