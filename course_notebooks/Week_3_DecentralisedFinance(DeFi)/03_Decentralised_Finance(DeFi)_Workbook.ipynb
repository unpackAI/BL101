```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#plt.style.use('_mpl-gallery')
import json
import requests
import datetime
```

```python
chart_data={}
# scrape M1  
try: 
    data = pd.read_html("https://take-profit.org/en/statistics/money-supply-m1/")
    chart_data["M1 worldwide (B USD)"] = data[0]["Actual"].sum()
except:
    chart_data["M1 worldwide (B USD)"] = 83194.75600000002

# scrape cryptos
try:
    url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=500&page=1&sparkline=false"
    response = requests.get(url)
    r = response.text
    data = json.loads(r)

    df = pd.DataFrame(data)
    chart_data["Crypto Market Cap (B USD)"] = df["market_cap"].sum() /10**9 #in billions
    chart_data["Bitcoin Market Cap (B USD)"] = df[df["name"]=="Bitcoin"].market_cap.values[0] /10**9
    chart_data["Ethereum Market Cap (B USD)"] = df[df["name"]=="Ethereum"].market_cap.values[0] /10**9
    date = datetime.datetime.now().strftime("%Y-%m-%d")
except:
    chart_data["Crypto Market Cap (B USD)"] = 920.895355363 /10**9 #in billions
    chart_data["Bitcoin Market Cap (B USD)"] = 405.203221292 /10**9
    chart_data["Ethereum Market Cap (B USD)"] = 136.661252887 /10**9
    date = "2022-06-14"
```

```python
url = "https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=usd&days=max&interval=monthly"
response = requests.get(url)
r = response.text
data = json.loads(r)

df_btc = pd.DataFrame(data)

# convert the datetime:
#from datetime import datetime
#datetime.fromtimestamp(timestamp/1000)
```

```python
url = "https://api.coingecko.com/api/v3/coins/ethereum/market_chart?vs_currency=usd&days=max&interval=monthly"
response = requests.get(url)
r = response.text
data = json.loads(r)

df_eth = pd.DataFrame(data)
```

```python
# Plot  
# make data:
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_yscale('log')
ax.bar(["M1", "Cryptos", "Bitcoin", "Ethereum"],chart_data.values())
plt.title("Worldwide Supply of Money vs Cryptocurrencies (in billion USD)")
plt.suptitle(f"(data scraped: {date})")
plt.show()
```
