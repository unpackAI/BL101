### Send Money ğŸ’¸
>Of course we could simply use the **Send Button** in your wallet ğŸ¦Š. But where would all the fun be ğŸ¤ª, if we don't code from scratch ğŸ“

**For sending Eth between two blockchain accounts, you need:**   
1. the `private key` of the account you are sending *from* ğŸ‘ˆğŸ» from_account_pk
2. the `wallet address` of the account you are sending *to* ğŸ‘ˆğŸ» to_account
3. how much you want to send

```python
# SEND ETH BETWEEN TWO ACCOUNTS âœ…

#We need: 
#   a) the private key of the account you are sending FROM ğŸ‘ˆğŸ» from_account_pk
#   b) the wallet address of the account you are sending TO ğŸ‘ˆğŸ» to_account

# 1ï¸âƒ£ INPUT DATA
from_account_pk = private_key # ğŸ‘ˆğŸ»  change
to_account = wallet_address2 # ğŸ‘ˆğŸ»  change
amount = 0.1 # ğŸ‘ˆğŸ»  change

# 2ï¸âƒ£ STORE SOME INFOS FOR LATER
from_account = w3.eth.account.privateKeyToAccount(from_account_pk).address
from_account_balance_before = web3.eth.getBalance(from_account)
to_account_balance_before = web3.eth.getBalance(to_account)

# 3ï¸âƒ£ START TRANSACTION
#get the nonce.  Prevents one from sending the transaction twice
nonce = web3.eth.getTransactionCount(from_account)

#build a transaction in a dictionary
tx = {
    'nonce': nonce,
    'to': to_account,
    'value': web3.toWei(amount, 'ether'),
    'gas': 2000000,
    'gasPrice': web3.toWei('50', 'gwei')
}

#sign the transaction
signed_tx = web3.eth.account.sign_transaction(tx, from_account_pk)

#send transaction
tx_hash = web3.eth.sendRawTransaction(signed_tx.rawTransaction)
```

```python
# PRINTOUT TRANSACTION

# convert to Eth
from_account_balance_before_eth = float(web3.fromWei(from_account_balance_before, 'ether')) 
from_account_balance_after_eth = float(web3.fromWei(from_account_balance_after, 'ether')) 
to_account_balance_before_eth = float(web3.fromWei(to_account_balance_before, 'ether')) 
to_account_balance_after_eth = float(web3.fromWei(to_account_balance_after, 'ether')) 

# build table
df = pd.DataFrame(data={"Account Number": [from_account, to_account], "Before (ETH)": [from_account_balance_before_eth, from_account_balance_after_eth], 
                        "After (ETH)": [to_account_balance_before_eth, to_account_balance_after_eth]}, 
                    index=["from:", "to:"], 
                    columns=["Account Number", "Before (ETH)", "After (ETH)"])
df["Difference"] = df["After (ETH)"]-df["Before (ETH)"]

# print stats & table
print("Transaction Hash: ", web3.toHex(tx_hash))
print(f"Amount sent: {amount} Eth")
gas_payed = tx["gas"] * web3.toWei(tx["gasPrice"], 'wei')
print(f"Gas payed: {gas_payed} Wei")
df
```
