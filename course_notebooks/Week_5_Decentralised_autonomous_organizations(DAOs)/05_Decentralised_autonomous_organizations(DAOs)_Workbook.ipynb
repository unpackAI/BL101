#### Choose your type of DAO Governance

One of the most fundamental decisions that you will have to make is how your DAO will be governed. Like we covered within the coursebook, the voting mechanism to determine voting power heavily influences the kind of DAO you would like to create. Aside from a few other options, Aragon gives you 3 main options.

<center><img src="https://www.dropbox.com/s/wwvgx3ajwg5wfir/type_of_dao.png?raw=1" alt="dao_creation" width="600"/></center>

- **Company**: This DAO utilizes the transferrable token voting power mechanism. Essentially, the more governance tokens you have, the more influential you will become on the DAO's decisions. In this case, since tokens can be bought, influence can be bought as well.
- **Membership**: This DAO utilizes the one-member-one-vote voting mechanism. This is in line with the proof-of-personhood participation model that we discussed. However, since one member could create several wallets, this is not bullet proof.
- **Reputation**: This DAO also utilizes the non-tranferrable token voting power mechanism. While the amount of tokens you own, also represents the amount of influence you have, you are not able to buy or sell the token. The token is send or erased from your wallet based on your positive or negative impact on the DAO. This mechanism holds similar elements to the skin-in-the-game model that we discussed.

#### Create an ERC20 Governance Token

Unfortunately, you are not yet able to add your previously created ERC20 token as a government token to your new DAO. This function is not yet available within Aragon but is expected to be launched in 2022 Q4.

Therefore, you will have to create an entirely new ERC20 token that will be used for governance, however you can store your previously created token within the vault of your DAO.

Go through the rest of the process to launch your DAO. If you need any help you may also inquire the [Aragon Documentation](https://documentation.aragon.org/products/aragon-client/how-to-create-a-dao-using-aragon-client).

Your startup should be based on a **Blockchain Business Model**. A business model describes a plan or strategy of a company to sell a product or service and earn profit from there. Example categories for Blockchain Business Models are:

||Business Model|Description|
|--|--|--|  
|1|Token Economy|offer utility tokens for various purposes and as investment incentive|
|2|BaaS (BC as a Service)|provide BC related services and support for other companies or DAOs|
|3| Development Platforms|provide a platform for BC technology| 
|4|BC Based|develop and sell BC related software and products|
|5|Network Fee Charge|provide BC network and charge usage fee|
|6|BC Professional Service|all kind of professional service related to BC: dAPP development, consulting, game designing, auditing, marketing, etc.|
|7|P2P BC Business Model|help BC users to interact with each other (peer-to-peer) and get paid with tokens or transaction fee. Examples are Filecoin, IPFS, etc.|

A blockchain business model has all the three main characteristics of blockchain technology: it is decentralized, based on peer-to-peer transactions, within a trusted and reliable network. 

*You can use the framework below, adapt it / enhance it / alter it to meet your project goal.* 🏗👷🏻‍♂️🚧

#### 3️⃣.1️⃣ Value Model 🏆  
>In the value model we want to answer a few core questions:  
- **Core Problem:** *What's the problem your business is solving?*  
- **Value Proposition:** *What core product benefit does your product bring to what customer segment?*   

#### 3️⃣.2️⃣ Application Model  ⛓⚙️
>In the Blockchain Model we want to understand how the protocol connects the value model to everything else. 
- **Application Layer:** *How does your blockchain based application work?*
- **Blockchain Layer:** *On what chain is your application built on and why (think of the Blockchain Trilemma)? Possibly consider Layer 2 chains.*
- **Token Economy:** *Do you have a native token as part of your application? Why and what is its functionality?*

#### 3️⃣.3️⃣ Distribution Model 🚚
>Here we want to understand how you would distribute the product to your community.
- **Target Community:** *Who is your target user?*
- **Community Distribution:** *How do you plan to engage and attract people to join your community (think of NFT whitelists for the first 100 users or airdrops etc.)?*
- **Partner Network:** *What partners will you collaborate with to increase distribution channels and gain legitimacy?*

#### 3️⃣.4️⃣ Economic Model 📊📈 
>Here you should present how you and your stakeholders benefit.  
- **Business Model:** *How does your native token appreciate in value and how do you make money?*
