## Introduction into GameFi/SocialFi
This week we will be diving into the newly developing field of blockchain application: GameFi/SocialFi. Let's start with defining the two fields.

> **GameFi (GameFinance)** involves blockchain games that offer economic incentives to play them, otherwise known as play-to-earn games. Typically, players can earn in-game rewards by completing tasks, battling other players or progressing through various game levels. Additionally, the player is able to own game assets in form of NFTs that can evolve and develop throughout the game.

> **SocialFi (SocialFinance)** is a combination of social networking and finance to create decentralized social applications. Typically, network participants interact through social challenges, through which they can earn in-game rewards.

Arguably, this field will offer a new cycle of Blockchain innovation beyond DeFi, and possibly initiate mass adoption to attract an entirely new wave of users.

However, while GameFi/SocialFi shows promise, recent examples have branded this industry to be more of a ponzi scheme rather than a new and exciting developing field. As such, we will critically address the issues the industry have faced while also outlining the potential for growth.

It is important to note that GameFi and SocialFi are categories on their own and should be differentiated. Currently however, a lot of projects are intertwined in either category which makes it difficult to fully seperate them. We expect that the differences will become more clear and distinct as the fields develop. Lets' dive.

### The Token Economy of GameFi projects.

#### The basic principles of a Game economy

A game‚Äôs economy is the flow of resources between its systems and players. You move through the game, complete tasks and win challenges through which you are rewarded with the in-game currency which can be used to buy items. Sounds familiar? Well thats because in-game economies have existed for a long-time. What is new to crypto games however, are the core concepts of decentralisation and autonomy. As a player you are not simply investing your time for the entertainment, but also have the chance to own part of the game yourself by owning game native tokens or in-game NFT items. Either can be exchanged for real money at any point, which makes a functioning token economy critical for a games success.

When talking game economies there are two helpful concepts: Taps and sinks.

A tap is the source of in-game resources (i.e. you receive gold for finishing quests). A sink is where resources are destroyed or exchanged (i.e. buy an item for the gold earned). The analogy of a tap and a sink makes a lot of sense because it allows us to build a mental picture of a resources flow; If the tap is generating the resource quicker than the sink can remove it then an excess builds. Any resource that is in excess becomes effectively worthless to the player and they stop interacting with the tap.

However, the inverse can also be true: When there is too little resource coming from a tap then there is little motivation for a player to interactive with it. The correct balance between a tap and a sink is known as a pinch point.

In economics a pinch point is the level of supply of a commodity at which demand is maximised due to consumers becoming concerned about the supply. In games we can think about this as the point where the supply shows the utility of a resource but is scarce enough to have players continue to seek it.

<img src="https://www.dropbox.com/s/qhn35ctoxqur9jy/Tap-and-Sink-In-Balance.jpeg?raw=1" alt="etherscan_week1" width="300"/>

#### Play-to-earn

The way crypto games have tried to achieve a pinch point, while still attracting crypto native players, is through the play-to-earn model. In order to enter and play-to-earn, players have to make an initial investment to qualify. This normally comes in form of an NFT, that costs around $100 to $1k (prices can vary due to volatility of coin prices), that differ based on their in-game traits and utility. Some might be more expensive due to a more variable trait such as being able to generate more resources more quickly. Now you can start earning.

Such model works if the player base continues to rise, and earned currency is reinvested into the economy, instead of simply being earned-and-dumped (exchanged for ETH or another currency outside the game's reach). If however, there are no new incoming players to purchase NFTs to play-to-earn and the existing playerbase is not reinvesting into the economy, the result will be an inflationary game economy.

#### Inflation & Deflation

Just like in the real economy, **inflation** and the abundance of in-game resources lead to an increase in cost of items. Additionally, the value of the token falls. For a player base that at current mostly consists of crypto investors that are driven by earn first, and put no emphasis on if the game is ‚Äúfun‚Äù such outcome would be deadly.

If the game provides enough incentive to reinvest and burn the players tokens in the process, the supply of overall coins is reduced and can increase price and so benefit the remaining holders. As such, token holders often like token burns when there is a set supply cap, making the token **deflationary**.

#### Multi-token economy

A game economy usually contain more than one in-game token that has one of the following functions (if you are able to discover new exciting projects with other token functionalities, let us know). These functions could be represented by: 

- **Governance Tokens**: These tokens act much like company shares, with the holder able to use their tokens to vote on gameplay and to gain a share of a profit.
- **Utility Tokens**: A token which can be exchanged for some kind of utility, such as craftng items, increasing speed of production etc.
- **Non-Fungible Tokens (NFTs)**: A token with unique metadata that is one of a kind. This token may also grant governance or utility aspects too, such as the ability to earn and or partake in competitions.

Without further redo, lets explore existing game economys.

#### On-chain Gaming
Blockchain games are games that have some of their core mechanics handled by the blockchain. It‚Äôs a continuum ranging from **"weakly" on-chain games** that for example only require the player to buy an NFT to play the off-chain game, to **"strongly" on-chain games** that require the player to interact with the blockchain in every detail of the game.

Running a game completely on-chain is difficult to be economically profitable. That is why most of the blockchain games that have been released in the past few years are in a hybrid form, having only a few components of their stack on-chain while running the core part of their game logics on off-chain proprietary servers.  

##### "Weakly" on-chain  
For weakly on-chain games the game logic is off-chain. Interoperability is very stricted and only possible between game assets. These game assets are often NFTs representing e.g. a character or a weapon:

*The advantages of weakly on-chain games still outnumbers the disadvantages:* 

**Advantages and Disadvantages of on-chain gaming**

The biggest downside with weakly on-chain gaming is that the game developer or publisher acts as a central authoritiy and could start censoring, changing rules, increase prices or disappeare. Hence, the concept of *play-to-earn* is finally in the hands of the game publisher.

On the good side, the user experience is quite similar to traditional games, and connection time is not a problem. Updates (e.g. to fix bugs) can be done centrally and players who play against the terms of services can be banned quickly.


##### "Strongly" on-chain  
**Strongly on-chain** games store all assets, skills, game and building logic, etc. fully on the blockchain. Every game component is stored in smart contracts and therefore potentially composable, interoperable, and infinitely duplicable. These components can be NTFs representing skill trees or *quest factory contracts* where anyone can deploy and import their game missions.   
Like for all games where the source code is open and the game logic revealed, a certain kind of randomness is necessary. ##### Challenges in on-chain GameFi and how it might be solved <a name="cite_ref-CHE"></a>[<sup>[CHE]</sup>](#cite_note-CHE)
>GameFi comes with excitement and skepticism: record-breaking fund-raising, play-to-earn business model, more control for players in the upcoming metaverse, etc. on the one side. And still unsolved issues related to on-chain gaming on the other side. 

*In the following section we list down some issues which are still not solved and offer perspectives on how to solve them.*


|Challenge|Overview|Solution|
|--|--|--|
|**Multi-accounting**|Mitigate Sybil attacks ruining the gaming experience|1. NFT-based access, 2. Proof-of-humanity, 3. Reputation system |
|**Key Management**|Balancing UX and security requirements|1. Local & temporary keys, 2. Smart wallets & multisignature|
|**Hiding Information**|Obscuring private gameplay information on a public database|1. Commit-reveal schemes, 2. Zero-knowledge proofs|
|**Cheat-to-Earn**|Hack Plack-to-Earn games to earn more rewards|1. Mitigate botting, 2. Decentralized anti-cheat, 3. Reputation system|

- **Multi-accounting** refers to how to identify the player. *Most games implicitly rely on the following rule: one player = one human. If not, this means that someone else could be playing in the name of the player*. Possible solutions are:   
    - An NFT acts as a proof of ownership. The NFT should also be costly enough so that "hijacking" another players account is not worth it.
    - To proof a player's human identity can come in many ways: a player can proof her identity by performing specific actions (e.g. minimum number of votes) or the game can have a built-in reputation system for every player (e.g. reputation increases with total time played). Finally, some third-party services (e.g. [Proof of Humanity](https://www.proofofhumanity.id/) offer on-chain identity-checking.  

- **Key Management** refers to the use of private keys to sign transactions. *The more actions players can perform, the more they will have to sign transactions, the more the Metamask wallet will pop up and annoy the player.* Possible solutions are:   
    -   key pairs for "non-sensitive" game actions stored temporary (e.g. during game session) local or on a client side 
    - use of **smart wallets** in which players can define in advance which kind of transactions are allowed

- **Hiding Information** refers to the problem that every player can look up the others players move on the blockchain. *Games like poker would be useless on-chain*. Possible solutions are:
    - Commit-reveal schemes use hashing to store "game-sensitive" information on-chain. E.g.: for a game like **rock-scissor-paper** each player would hash his choice together with a number or word. After both players transmitted the hash they have to reveal their number to proof the correctness of their choice.
    - Zero-knowledge proofs store only the result of a set of actions on-chain rather than the actions itself (e.g. a cook does not need to reveal it's recipe to prove that he can cook a certain meal or a sudoku solution can be proofed by summing up rows and columns).    

- **Cheat-to-earn** refers to the problem that players could use *game-bots* which plays in their behalf. This is not pure on-chain gaming problem as more a problem in play-to-earn settings. Possible solutions are:
    - make the game settings too difficult for bots  
    - implement human interaction (e.g. players need to interact ocassionally on *Discord*)   
    - reward player with a "clean" cheating history or implement a strict rule that any cheating will result in banning.

>*These are all problems that hasn't been solved as a whole by any blockchain game so far. However, some games already implemented solutions to these challenges and further progress can be expected. In the mean-time *weakly-chained gaming* will characterize most games. Also the block time and overall blockchain speed is a problem until blockchains have sped up.*

### StarkNet 
[StarkNet](https://starkware.co/starknet/) is a good example of a  **Zero-knowledge rollup** (ZK-rollup). It operates as an L2 network over Ethereum, enabling any dApp to achieve unlimited scale for its computation ‚Äì without compromising Ethereum's composability and security.  
ZK-rollups are Layer 2 blockchain networks which are derrived or inherited from layer 1 blockchains like Ethereum.  
ZK-rollups use a mathematical construct called *Merkle Tree* (or hash tree üëâüèª week1: *like block hashing but in a tree structure*): a zk-rollup consists of two Merkle Trees which are both stored on-chain. One tree is dedicated to store accounts, while the other stores all balances. Any other type of data generated and used by the zk-rollup is stored off-chain.

One of the reasons why zk-rollups manage to perform significantly better than Layer 1 counterparts is the off-chain storage of data. Merkle Trees only consist of the most important data relevant to the smart contract and are accessed and requested to output information much more infrequently than in Layer 1 solutions. This saves huge amounts of processing power and time for the blockchain. 

With a ZK-rollup, validating a block is quicker and cheaper because less data is included. It's also secure and decentralized, since the data that is needed to recover the state is stored on the layer 1 chain (Ethereum).

Zk-rollups are becoming increasingly popular both among developers who are looking to boost usability, and among investors and traders looking for faster and cheaper transaction processing.

### SocialFi Development
>SocialFi is social influence on any medium that can be tokenized, allowing anybody to get direct monetary benefits from being on SocialFi platforms.

The center of SocialFi is NFTs. NFTs make it possible to securely trade and sell digital content. They enable digital proof of ownership in real-world experiences that brands can create for their communities. Brands can generate revenues from these experiences after-the-fact, inspiring them to create viral content in the metaverse. E-Commerce will be extended by a virtual dimension in the metaverse leading to a marketing section: **virtual commerce** (V-Commerce):
V-Commerce refers to a blend of digital and physical commerce. Unlike E-Commerce, V-Commerce will build a more seamless mix of digital and physical realities.

### GameFi Development  
If we count *CryptoKittens* as the first example of how gaming could be introduced onto blockchains in November 2017 - *almost 2 years after Ethereum was introduced* - then we are as of today already 5 years into the GameFi age. 

The number of GameFi projects has boomed in 2021 and will likely continue to grow in the coming years. As of March 2022, there are more than 1,400 blockchain games listed in DappRadar with popular games across multiple blockchains, such as Ethereum, BNB Smart Chain (BSC), Polygon, Harmony, Solana, and many more.

As blockchain technology continues to develop, the GameFi growing trend is expected to continue at a fast pace. The ability to own in-game assets and make money from games makes GameFi very attractive, especially in developing countries.

The current GameFi landscape consists mostly of "weakly" on-chain games with a focus on *play-to-earn*:  in-game rewards can be transacted as NFTs on crypto exchanges and marketplaces, allowing players to cash them out into other cryptocurrencies and ultimately fiat.  The actual gameplay however, is (still) mostly a second priority making current web3 games being dis-engaging and tedious to play with. 

Although static games like *Axies Infinity* still hold the majority of the GameFi market we are starting to see the rise of higher quality metaverse games such as *Decentraland* and *The Sandbox*.

A typical **triple A** game takes years to be built. For example the famous *Grand Theft Auto* by Rockstar took 3 years to develop and with smart contracts being relatively new, it is expected that a true triple A game is still at least a couple of years away.

In general we might see more advanced on-chain games in which not only *play-to-earn* but more and more *play-for-fun* content dominates.
